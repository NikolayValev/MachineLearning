Import the dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import decomposition
from sklearn import preprocessing
from sklearn import metrics
from sklearn.preprocessing import scale

%matplotlib inline
plt.style.use('seaborn-white')
```

(a)  Given the provided datasets (as CSV files), load them in Python.

```python
# Load Data and Inspect
NCI60_data = pd.read_csv('NCI60_data.csv', index_col=0)
NCI60_labs = pd.read_csv('NCI60_labs.csv', index_col=0)
NCI60_labs.x
```

Data preprocessing:

```python
# normalize the data attributes
normalized = preprocessing.normalize(NCI60_data)
# fit and transform the data
standardized = preprocessing.scale(NCI60_data, axis=0)
```

Checking the mean and the standard deviation

```python
standardized.mean(), standardized.std()
```

(c)
i
Fit the PCA model and transform to get the principal components

```python
from sklearn.decomposition import PCA
pca_data = PCA().fit_transform(standardized)
print(pca_data[:,0])
```

```python
col_names=[]
for x in range(len(NCI60_labs.x)):
  col_names.append("PC"+str(x+1))
principal_components = pd.DataFrame(
    pca_data, columns=col_names, index=NCI60_labs.x)
principal_components
```

(c)
ii 
Plot the first few principal component score vectors (i.e.Z1 vsZ2 andZ1vsZ3), in order to visualize the data. Comment on your visualization.Note: The observations corresponding to a given cancer type should beplotted in the same color, so that we can see to what extent the observa-tions within a cancer type are similar to each other.

```python
#get the array of random colors
from random import randint
colors = []

for i in principal_components.index:
    colors.append('#%06X' % randint(0, 0xFFFFFF))
```

```python
fig, axs = plt.subplots(figsize=(9, 7))
# Plot Principal Components 1 and 2
for i in principal_components.index:
  axs.annotate(i,(
               principal_components.PC1.loc[i], -
               principal_components.PC2.loc[i]),
               ha='center', color=colors[1])
  axs.annotate(i, (principal_components.PC1.loc[i],
               - principal_components.PC3.loc[i]),
               ha='center', color=colors[2])
  axs.annotate(i,(
               principal_components.PC2.loc[i], -
               principal_components.PC3.loc[i]),
               ha='center', color=colors[3])
```

```python
fig, ax1 = plt.subplots(figsize=(9, 7))

ax1.set_xlim(-3.5, 3.5)
ax1.set_ylim(-3.5, 3.5)

# Plot Principal Components 1 and 2
for i in principal_components.index:
    ax1.annotate(
        i, (principal_components.PC1.loc[i], -principal_components.PC2.loc[i]),
         ha='center')

# Plot reference lines
ax1.hlines(0, -3.5, 3.5, linestyles='dotted', colors='grey')
ax1.vlines(0, -3.5, 3.5, linestyles='dotted', colors='grey')

ax1.set_xlabel('First Principal Component')
ax1.set_ylabel('Second Principal Component')

# Plot Principal Component loading vectors, using a second y-axis.
ax2 = ax1.twinx().twiny()
ax2.set_ylim(-1, 1)
ax2.set_xlim(-1, 1)
ax2.tick_params(axis='y', colors='orange')
for i in principal_components.index:
  ax1.annotate(i,(
               principal_components.PC1.loc[i], -
               principal_components.PC2.loc[i]),
               ha='center', color='blue')
```

Plot the Proportion of Variance Explained (PVE) of each principal component 
(i.e. a scree plot) and the cumulative PVE of each principal component.

```python
for i in len(principal_components): {
    print("Standard Deviations of Principal Component: \n",
          np.sqrt(pca_data[i].explained_variance_))

}

# Percentage of Variance Explained (PVE) by each PC:
plt.figure(figsize=(6, 4))
plt.ylim(0, 1.05)

plt.plot(pca_data.explained_variance_ratio_,
         '-o', label='Individual component')
plt.plot(np.cumsum(pca_data.explained_variance_ratio_),
         'bo-', label='cumulative PVE')

plt.ylabel('percentage of variance explained')
plt.xlabel('Principal components')
plt.legend(loc='best')
plt.tight_layout()
```
